MULTICLASS CLASSIFICATION USING SVM ON IRIS DATASET

```python
# Libraries Installation
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,mean_absolute_error,mean_squared_error
import sklearn.model_selection as model_selection
from sklearn.metrics import classification_report,confusion_matrix
from sklearn import preprocessing
```

```python
mydata = pd.read_csv(r"D:\1st semester\DA lab data sets\Iris.csv")
```

```python
mydata
```

```python
pd.get_dummies(mydata,columns=['Species'],drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(mydata.drop('Species',axis=1))
scaled_features = scaler.transform(mydata.drop('Species',axis=1))
mydata_scaled = pd.DataFrame(scaled_features,columns=mydata.columns[:-1])
mydata_scaled.head(10)
```

```python
x=mydata_scaled
y=mydata['Species']
```

```python
x=mydata.drop(labels='Species',axis=1)
y=mydata['Species']
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.25, random_state=50)
```

```python
y_test.value_counts()
```

```python
svc_model = SVC()
svc_model.fit(x_train,y_train)
```

```python
y_predict = svc_model.predict(x_test)
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
print(classification_report(y_test, y_predict))
```
